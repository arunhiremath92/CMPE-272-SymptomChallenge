### Calculate number of days it has been since policy was last implemented

```python
policy_list = ["School_close_recommended","School_close_reqd_some_lvl","School_Closure_All_Levels",
               "Workplace_Closure_Recommended","Workplace_close_reqd_some_lvl","Workplace_Closure_Required",
              "Cancel_PE_recommended","Cancel_PE_required",
               "Testing_policy_SC","Testing_policy_symptomatic","Testing_policy_drive_through",
               "Stay_at_home_recommended","Stay_at_home_exceptions","Stay_at_home_required",
              "Facial_Covering_recommended","Facial_covering_reqd_lvl1","Facial_covering_reqd_lvl2","Facial_covering_reqd_total"]
nat = np.datetime64('NaT')
#frame

frame.sort_values(['Date'], inplace=True)
fr1 = frame.copy()

for i in range(0, 18):
#     print(i)
    datevalue = pd.DataFrame(state_df.where((state_df.RegionName.isin(frame['Province_State'])) 
                        & (state_df['Policy ' + str(i + 1)] == 1)
                        & (state_df['Policy ' + str(i + 1)].shift() == 0 )
                      )[['Date', 'RegionName']]).groupby('RegionName')['Date'].apply(list).reset_index(name='values')
    temp = datevalue.copy();
    temp = temp.explode('values').reset_index(drop=True)
    temp['cDate'] = temp['values'].astype('datetime64[ns]')
    fr = frame.copy()
#     fr = fr[frame.Province_State != '']
    fr['cDate'] = fr['Date'].astype('datetime64[ns]')
    diff = []
    pDate = []
    for row in fr.itertuples():
        t = temp[temp.RegionName.eq(row.Province_State)]
        date = t.where(t.cDate.notnull() & t.cDate.lt(row.cDate)).sort_values(by=['cDate'], ascending=False).cDate
        if (date.values.size != 0):
            if(np.isnat(date.values[0]) == False):
                diff.append((row.cDate - date.values[0]))
                pDate.append(date.values[0])
            else:
                diff.append(pd.NaT)
                pDate.append(pd.NaT)
#             print(row.Province_State)
#             print(row.cDate)
#             print(date.values[0])
        else:
            diff.append(pd.NaT)
            pDate.append(pd.NaT)
    fr1['policy_date' + policy_list[i] ] = pDate
    fr1[policy_list[i]] = diff


fr1
```
