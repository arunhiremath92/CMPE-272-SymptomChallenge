## Read Data File - COVID ACT NOW

```python
df = pd.read_csv('coviact_RECENT.csv')
df
```

## Fill 0 values for NaN

```python
df['actuals.cases'].fillna(0, inplace=True)
df['actuals.newCases'].fillna(0, inplace=True)
df['actuals.deaths'].fillna(0, inplace=True)

df['actuals.positiveTests'].fillna(0, inplace=True)
df['actuals.positiveTests'] = df['actuals.positiveTests'].astype('int64')

df['actuals.negativeTests'].fillna(0, inplace=True)
df['actuals.negativeTests'] = df['actuals.negativeTests'].astype('int64')

df['metrics.testPositivityRatio'].fillna(0, inplace=True)
df['metrics.caseDensity'].fillna(0, inplace=True)
df
```

## Preprocess Data and remove insignificant values

```python
wpt = df[((df['actuals.positiveTests'] > 0) | (df['actuals.negativeTests'] > 0))&
        (df['metrics.caseDensity'] > 0)]
wpt
```

```python
# wtpr = wpt[wpt['metrics.testPositivityRatio'] > 0]
# wcd = wtpr[wtpr['metrics.caseDensity'] > 0]
```

## Derive cumulative testing volume 

```python
# totalTestingVolume -> Aggregated test volume to date
wpt['totalTestingVolume'] = wpt.apply(lambda row: row['actuals.positiveTests'] + row['actuals.negativeTests'], axis=1)
wpt
```

### Derive daily testing volume

```python
import numpy as np
wpt.sort_values(['date', 'fips'], inplace=True)

# testingVolume -> Daily testing volume
wpt['testingVolume'] = wpt.groupby(['fips'])['totalTestingVolume'].diff().fillna(0)
wpt.sort_index(inplace=True)

# Convert date column to date type to enable comparison
wpt['date'] = wpt['date'].astype('datetime64[ns]')
# Get only data where testing volume is available
# '2020-08-03' - Earliest date for which we have testing volume data per county
wpt = wpt[wpt['date'] > '2020-08-03']
wpt
```

```python
# Filter out daily testingVolume < 0. This happens because of faulty or incomplete data in some cases.
# For eg. St.Helena Parish county in early November
wpt = wpt[wpt['testingVolume'] > -1]
```

## Compute state median for testing volume by date

```python
wpt['date'] = wpt['date'].apply(lambda x : x.date())
grp = wpt.groupby(['date', 'state'])['testingVolume'].apply(np.median).reset_index()
grp['stMedian'] = grp['testingVolume']
grp
```

### Join state median testing volume with cleaned data

```python
wpt['date'] = wpt['date'].astype('datetime64[ns]')
mrg = wpt.merge(grp, left_on=['state', 'date'], right_on=['state', 'date'], how='inner')
mrg
```

```python
mrg['testingVolume'] = mrg['testingVolume_x']
mrg
```

### Drop extraneous columns

```python
mrg = mrg.drop(columns=['testingVolume_x', 'testingVolume_y'])
mrg
```

### Define helper functions for computing adjusted case rate, and county tier

```python
def getAdjustedCaseRate(test_positivity, test_vol_upon_st_median, case_rate, population):
    adjusted_case_rate = case_rate;
    if population < 106000:
        return adjusted_case_rate
    else:
        volume_upon_median = test_vol_upon_st_median
        if (volume_upon_median > 1 ):
            if(volume_upon_median >= 2.00):
                adjusted_case_rate = 0.5*case_rate
            elif(volume_upon_median >= 1.75):
                adjusted_case_rate = 0.625*case_rate
            elif(volume_upon_median >= 1.5):
                adjusted_case_rate = 0.75*case_rate
            elif (volume_upon_median >= 1.25):
                adjusted_case_rate = 0.875*case_rate

        elif (volume_upon_median < 1 and test_positivity > 0.035):
            if (volume_upon_median >= 0.75):
                adjusted_case_rate = 1.1*case_rate
            elif(volume_upon_median >= 0.50):
                adjusted_case_rate = 1.2*case_rate
            elif(volume_upon_median >= 0.25):
                adjusted_case_rate = 1.3*case_rate
            elif(volume_upon_median >= 0.00):
                adjusted_case_rate = 1.4*case_rate
    #print(adjusted_case_rate)
    return adjusted_case_rate
```

```python
def getTierForCounty(test_positivity, adjusted_case_rate):
    percent_number = test_positivity * 100
    tpr = round(percent_number, 1)
    if(tpr > 8):
        tier_tpr = 1
    elif (tpr >= 5):
        tier_tpr = 2
    elif (tpr >= 2):
        tier_tpr = 3
    else:
        tier_tpr = 4
    
    acr = round(adjusted_case_rate, 1)
    if(acr > 7):
        tier_acr = 1
    elif (acr >= 4):
        tier_acr = 2
    elif (acr >= 1):
        tier_acr = 3
    else:
        tier_acr = 4
    #print(tier_acr, tier_tpr)
    #print(min(tier_tpr, tier_acr))
    return min(tier_tpr, tier_acr)
```

```python
def getOverallTier(county):
    population = county['POPESTIMATE2019']
    adjusted_case_rate = getAdjustedCaseRate(county['metrics.testPositivityRatio'],county['testingVolProportion'],county['metrics.caseDensity'],population) 
    test_positivity_rate = county['metrics.testPositivityRatio']
    return getTierForCounty(test_positivity_rate,adjusted_case_rate)
```

```python
mrg2['tier'] = mrg2.apply(lambda x: getOverallTier(x), axis=1)
mrg2
```
